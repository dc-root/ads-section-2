Estou interessado nos passageiros do Titanic das classes de cabine 2 e 3.

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]

class_23.head()
```

Semelhante à expressão condicional, a isin()função condicional retorna um Truepara cada linha em que os valores estão na lista fornecida. Para filtrar as linhas com base nessa função, use a função condicional dentro dos colchetes de seleção []. Nesse caso, a condição dentro dos colchetes de seleção verifica para quais linhas a coluna é 2 ou 3.titanic["Pclass"].isin([2, 3]) Pclass

O acima é equivalente a filtrar por linhas para as quais a classe é 2 ou 3 e combinar as duas instruções com um |operador (or):

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]

class_23.head()
```
